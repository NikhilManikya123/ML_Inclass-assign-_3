```python
#1B Solution
arr2 =np.array([[3,-2],[1,0]])
eigenval,eigenvect = np.linalg.eig(arr2)
print("Eigenvalues:",eigenval)
print("Right eigenvectors:")
print(eigenvect)
```

```python
#1C Solution
arr3= np.array([[0,1,2],[3,4,5]])
diag_sum=np.trace(arr3)
print("Diagonal sum:", diag_sum)
```
